### Importing libraries

```python
import requests
from skimage import io
```

### API key and base URL for the NASA APIs

```python
api_key = 'DEMO_KEY'
api_url = 'https://api.nasa.gov'
```

## Build the request URL for the APOD API

```python
apod_url = f'{api_url}/planetary/apod'
apod_params = {'api_key': api_key}
```

### Send a GET request to the APOD API endpoint

```python
apod_response = requests.get(apod_url, params=apod_params)
```

### Check if the request was successful

```python
apod_response.status_code == 200
```

### Parse the response data as JSON

```python
apod_data = apod_response.json()
```

### Extract the APOD title and explanation from the response data


```python
apod_title = apod_data['title']
apod_explanation = apod_data['explanation']
img_url = apod_data['url']
```

### Print the APOD title and explanation


```python
print(f'Astronomy Picture of the Day: {apod_title}\n')
print(apod_explanation)
```

### Display the APOD image using scikit-image library

```python
img = io.imread(img_url)
io.imshow(img)
```

## Build the request URL for the Mars Rover Photos API

```python
rover_name = 'curiosity'
rover_url = f'{api_url}/mars-photos/api/v1/rovers/curiosity/photos'
params = {'sol': '1000', 'api_key': api_key}
```

### Send a GET request to the Mars Rover Photos API endpoint

```python
response = requests.get(rover_url, params= params)
```

### Check if the request was successful

```python
response.status_code == 200
```

### Parse the response data as JSON

```python
data = response.json()
```

### Extract the photos from the response data

```python
photos_data = data.get('photos')
```

### Print the full name of the camera


```python
print('\nPhotos taken by the Mars Curiosity Rover:')

for photo_data in photos_data:
    print(photo_data['camera']['full_name'])
```

### Display one of the photos using the scikit-image library

```python
img_url1 = photos_data[0]['img_src']
img1 = io.imread(img_url1)
camera_name = photos_data[0]['camera']['full_name']
io.imshow(img1)
print(f"Camera used: {camera_name}")
```
